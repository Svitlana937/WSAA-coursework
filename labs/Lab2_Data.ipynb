```python
import csv
```

```python
FILENAME= "data.csv"
DATADIR = ""

with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    for line in reader:
        print (line) # list type
```

```python
with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    linecount = 0
    for line in reader:
        if not linecount: # first row ie header row
            print (f"{line}\n-------------------")
        else: # all subsequent rows
            print (line)
        linecount += 1
```

```python
# 4.a
with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    linecount = 0
    total = 0
    for line in reader:
        if not linecount: # first row ie header row
            pass
        else: # all subsequent rows
            total += int(line[1]) # why 1
        linecount += 1
    print (f"average is {total/(linecount-1)}") # why -1 ? Answer: We adding -1 beacuse we want to make our calculation without header row
```

```python
# 4.b
with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",", quoting=csv.QUOTE_NONNUMERIC)
    linecount = 0
    total = 0
    for line in reader:
        if not linecount: # first row ie header row
            pass
        else: # all subsequent rows
            total += line[1] # why 1
            linecount += 1

print (f"average is {total/(linecount-1)}") # Answer: we will get an error on Devision by zero.
```

```python
# 5.

with open (DATADIR + FILENAME, "rt") as fp:
    # DictReader сам понимает, что первая строка — это заголовки
    reader = csv.DictReader(fp, delimiter=",", quoting=csv.QUOTE_NONNUMERIC)
    total = 0
    count = 0
    for line in reader:
        # Теперь мы обращаемся к данным по имени колонки 'age'
        total += line['age']
        # print (line)
        count += 1

print (f"average is {total/(count)}") # why is there no -1 this time? The Answer: if we will add -1 it will count the header also.
```
